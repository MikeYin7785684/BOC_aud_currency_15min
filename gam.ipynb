```r
mcmc_stop<- function(net,method,iteration,parameter1,parameter2){
    # '''method is a funntion of random variable eg. rnorm and paramters is rnorm(1,mean=parameter1,variance=paramter2)'''
    A<- NULL
    for (i in 1:iteration) {
        a <- 0
        j<-0
        while (a < net) {
            a <- a + method(1, parameter1, parameter2)
            j<-j+1
        }
        A <- c(A,j )
    }
    A
}
par(bg='white')
plot(hist(mcmc_stop(1, rbinom, 10000, 1, 0.6)))
trial<-NULL
for (i in 1:1000) {
    trial <- c(trial, mean(mcmc_stop(1, rbinom, 1000, 1, 0.6)))
}
plot(density(trial))
plot(dbinom(1:1000,1000,0.9))
```

```r
D_mnp<-function(m,n,p){
    n_m<- n-m
    if(m<=n){
        if (n==1){
            return(p)
        }
        else{
            if(m>n_m){
                return(D_mnp(n_m,n,p))
            }
            else {
                if(m==0){
                    return(D_mnp(m,n-1,p)*(1-p))
                }
                else {
                    return(D_mnp(m-1,n-1,p)*p+D_mnp(m,n-1,p)*(1-p))
                }  
            }
        }    
    }
    else {
    print('error:m>n')
    return(0)
    }
}
```

```r

randomwalk_stop <- function(m,b,n,p){
    if(m>=b){
        x<-2*m-b
        cd<-n-x
        if(cd%%2 ==0 ){
            return(D_mnp(x,n,p))
        }
        else {
           print(('error: condition not satisfied'))
           return(0)
        }
    }
    else {
       print('error:m should larger or equal to n')
       return(0)
    }
}
```
